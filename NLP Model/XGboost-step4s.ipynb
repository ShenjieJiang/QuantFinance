```python
# 1 ~ 11 训练集一年，测试及半年
# 1 ~ 6 训练半年, 测试一年

X_train = feature[train_index3]
y_train = label[train_index3]

X_test = feature[test_index3]
y_test = label[test_index3]

# 超参数网格
param_grid = {'learning_rate': [0.025, 0.05, 0.075, 0.1],
              # 'n_estimators': [20, 30, 40, 50, 70],
              'max_depth': [3, 5, 7],
              'subsample': [0.8, 0.85, 0.9, 0.95],
              'seed': [42],
              'reg_alpha': [1e-1],       # 增加L1正则化项
              'reg_lambda': [1e-1],      # 增加L2正则化项
              'booster': ['gbtree'],
              'objective': ['multi:softprob'],
              'num_class' : [3],
              # 'tree_method': ['gpu_hist'],  # 使用GPU进行树的构建
              # 'predictor': ['gpu_predictor']  # 使用GPU进行预测
             }

# 创建XGBoost分类器，指定损失函数为多类别分类的Softmax
xgbcls = xgb.XGBClassifier()

grid_search = GridSearchCV(xgbcls, param_grid,
scoring='neg_log_loss', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
model = grid_search.best_estimator_

# 训练模型
model.fit(X_train, y_train)

# 进行预测
y_pred = model.predict(X_test)

# 计算分类准确度
accuracy_t = accuracy_score(y_train, model.predict(X_train))
accuracy_p = accuracy_score(y_test, y_pred)
print(f"Tran_Accuracy: {accuracy_t:.2f}")
print(f"Pred_Accuracy: {accuracy_p:.2f}")
```

```python
pred['prob_0'] = prob[:,0]
pred['prob_1'] = prob[:,1]
pred['prob_2'] = prob[:,2]
pred
```

```python
pred['L_l'] = np.log(pred['prob_0']/(1-pred['prob_0']))
pred['L_m'] = np.log(pred['prob_1']/(1-pred['prob_1']))
pred['L_h'] = np.log(pred['prob_2']/(1-pred['prob_2']))
pred['forecast_adj_txt'] = pred['L_h'] - pred['L_l']
pred
```

```python
mean_ = pred.groupby(['time', 'stock_code']).mean()
mean_ = mean_.reset_index()
mean_ = mean_.pivot(index='time', columns='stock_code', values='forecast_adj_txt')
mean_
```
